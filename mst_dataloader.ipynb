```python
df = pd.read_csv('train.csv')
df.head()
```

```python
# Split Train Val sets
df_train, df_val = train_test_split(df, test_size=0.2, random_state=42)
```

```python
df_train.reset_index(inplace=True,drop=True)
df_train = df_train[:1000] # just for test
df_train.head()
```

```python
df_val.reset_index(inplace=True,drop=True)
df_val = df_val[:1000] # just for test
df_val.head()
```

```python
def seperate_labels(dfrme):
  labels_df, imgs_df = dfrme.iloc[:,0], dfrme.iloc[:,1:]
  return imgs_df, labels_df
```

```python
# Create dataloader
def create_dl(dfrme, batch_size=128):
  x_train, y_train = seperate_labels(dfrme)

  X_train = torch.cat([torch.tensor(x_train.iloc[i,:]) for i in range(len(x_train))]).view(-1, 28*28) # stack each imgs, change shape to 28*28
  X_train = X_train.float()/255

  Y_train = torch.tensor(y_train).unsqueeze(1)

  dset = list(zip(X_train, Y_train))
  dl = DataLoader(dset, bs=64)

  return dl
```

```python
train_dl = create_dl(df_train)
```

```python
train_dl
```

```python
val_dl = create_dl(df_val)
```

```python
val_dl
```

```python
# create data loaders
dls = DataLoaders(train_dl, val_dl)
```
